```python
class Solution:
    def new21Game(self, n, k, maxPts):
        # Base case: the outcome is guaranteed
        if k == 0 or n >= k + maxPts - 1:
            return 1.0

        # dp[i] will store the probability of getting exactly i points
        dp = [0.0] * (n + 1)

        # Alice starts at 0 points with 100% probability
        dp[0] = 1.0

        window_sum = 1.0
        probability_sum = 0.0

        for i in range(1, n + 1):
            # The probability of reaching i is the average of the valid window
            dp[i] = window_sum / maxPts

            # If this state is sill < k, Alice can keep drawing from it.
            # It will contribute to future states.
            if i < k:
                window_sum += dp[i]

            else:
                # If i >= k, Alice stops drawing.
                # This is a final state; collect it if it's within our target n
                probability_sum += dp[i]

            # Slide the window forward: remove the sate that is now too far back
            # Only subtract if that old state was allowed to draw (i.e., < k)
            if i - maxPts >= 0 and i - maxPts < k:
                    window_sum -= dp[i - maxPts]

        return probability_sum
```

```python
solution = Solution()
```

```python
n = 10
k = 1
maxPts = 10

# Expecged: 1.0
print(solution.new21Game(n, k, maxPts))
```
