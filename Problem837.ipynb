```python
class Solution:
    def new21Game(self, n, k, maxPts):
        # Base case: the outcome is guaranteed
        # k = 0 -> Alice won't draw anything and her score starts with 0 points. -> 100%
        # Before the final draw Alice can have at most k - 1 points.
        # Then she can draw maxPts.
        # Maximum final score = (k - 1) + maxPts.

        if k == 0 or n >= (k - 1) + maxPts:
            return 1.0

        # dp[i] stores the probability that Alice is currently at i points
        # while she is still following the drawing process.
        dp = [0.0] * (n + 1)

        # Alice starts at 0 points with 100% probability
        dp[0] = 1.0

        # The sum of all probabilities from states that can still draw.
        # These states contribute to future probabilities.
        window_sum = 1.0

        # The probability that Alice ends the game with a final score in [k, n].
        probability_sum = 0.0

        for i in range(1, n + 1):
            # The probability of being at i comes from all previous valid states.
            # Since every draw has equal probability, divide their sum by maxPts.
            dp[i] = window_sum / maxPts

            # If this state is still < k, Alice can keep drawing from it.
            # It will contribute to future states.
            if i < k:
                window_sum += dp[i]

            else:
                # This is a final state because Alice stops drawing here.
                # Add its probability to the answer.
                probability_sum += dp[i]

            # Slide the window forward: remove the state that is now too far back
            # We added the second condition i - maxPts < k becasue
            # we only subtract if that old state was allowed to draw (i.e., < k)
            if i - maxPts >= 0 and i - maxPts < k:
                window_sum -= dp[i - maxPts]

        return probability_sum
```

```python
solution = Solution()
```
