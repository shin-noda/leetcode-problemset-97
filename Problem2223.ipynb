```python
class Solution:
    def sumScores(self, s):
        # Use Z-Algorithm
        # Idea: we calculate z-scores and store them in z-array
        # e.g. z[2] means how many characters from s[2:] match with the prefix of s.
        # Let's say s = "ababccee". Then, s[2:] = "abccee"
        # First two characters, "ab" match, so z[2] = 2.

        n = len(s)
        z = [0] * n

        # L and R form a window which matches a prefix of 's'
        # [l, r] is the window where we already know that s[l:r+1] matches the prefix s[0:r-l+1]
        # e.g. l=10, r=20 => s[10:21] == s[0:11]
        l, r = 0, 0

        for i in range(1, n):
            # If i is within the current matching window
            if i <= r:
                # Use previously computed Z-values to skip redundant matches
                # If I'm at position 'i' inside the window, then the matching
                # position inside the prefix is 'i - l'
                # 'i - 1' is a prefix position corresponds to 'i'.
                z[i] = min(r - i + 1, z[i - l])

            # Try to extend the match manually past the current window
            while i + z[i] < n and s[z[i]] == s[i + z[i]]:
                z[i] += 1

            # If the new match extends past 'r', update the window boundaries
            if i + z[i] - 1 > r:
                l = i
                r = i + z[i] - 1

        # The score of s_n (the full string itself) is always n
        return sum(z) + n
```

```python
solution = Solution()
```

```python
s = "babab"

print(solution.sumScores(s))
```

```python
s = "azbazbzaz"

print(solution.sumScores(s))
```
