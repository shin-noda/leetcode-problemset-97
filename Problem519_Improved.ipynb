```python
import random

class Solution:

    def __init__(self, m, n):
        # Use Fisher-Yates Shuffle.
        self.m = m
        self.n = n
        self.total_cells = m * n

        # Map: actual_index -> swapped_index
        self.mapping = {}


    def flip(self):
        # Pick a random index from the remaining available pool size
        self.total_cells -= 1
        rand_val = random.randint(0, self.total_cells)

        # Check if this index was previously swapped.
        # If not, its value is just itself.
        # Note: it's not a mistake that it's (rand_val, rand_val)
        actual_val = self.mapping.get(rand_val, rand_val)

        # Swap the picked element with the element at the end of the pool
        # (If the end element was already swapped, use its mapped value)
        # Note: it's not a mistake that it's (self.total_cells, self.total_cells)
        end_val = self.mapping.get(self.total_cells, self.total_cells)
        self.mapping[rand_val] = end_val

        # Convert linear index to 2D matrix coordinates
        row = actual_val // self.n
        col = actual_val % self.n

        return [row, col]


    def reset(self):
        # Clear the map and reset the pool size back to full in O(1) time
        self.mapping.clear()
        self.total_cells = self.m * self.n
```
