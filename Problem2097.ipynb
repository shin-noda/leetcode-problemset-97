```python
class Solution:
    def validArrangement(self, pairs):
        # User Hierholtzer's Algorithm
        # Start Node: out-degree = in-degree + 1
        # End Node: in-degree = out-degree + 1
        # All other nodes: in-degree = out-degree
        # If all nodes form a circle like 1->2->3->1...
        # then, all nodes can be a starting point, so
        # we just pick the first element of the first pair

        adj = {}
        node_map = {}
        for src, dst in pairs:
            if src not in node_map:
                node_map[src] = 0

            if dst not in node_map:
                node_map[dst] = 0

            node_map[src] += 1
            node_map[dst] -= 1

            if src not in adj:
                adj[src] = []

            adj[src].append(dst)

        # We pick this as default
        start = pairs[0][0]
        for node in node_map:
            if node_map[node] == 1:
                start = node
                break

        # This path is stored in a reverse order
        r_path = []

        def dfs(node):
            # Note: the base case is implicitly handled by the while loop condition.

            while node in adj and adj[node]:
                next_node = adj[node].pop()
                dfs(next_node)

            # Dead end reached
            r_path.append(node)

        dfs(start)

        path = r_path[::-1]

        result = []
        for i in range(len(path) - 1):
            result.append([path[i], path[i + 1]])

        return result
```

```python
solution = Solution()
```

```python
pairs = [[5,1],[4,5],[11,9],[9,4]]

print(solution.validArrangement(pairs))
```
