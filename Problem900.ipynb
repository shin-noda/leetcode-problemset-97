```python
class RLEIterator:

    def __init__(self, encoding: List[int]):
        # Idea: try implementing the lazy evaluation solution
        # instead of eager evaluation.
        self.encoding = encoding
        self.m = len(self.encoding)

        # Note: this index points, 0, 2, 4, ..., 2k, only even numbers
        self.index = 0
        self.remaining = self.encoding[0]


    def next(self, n: int) -> int:
        if self.index >= self.m:
            # No element to exhaust
            return -1

        self.remaining -= n
        if self.remaining >= 0:
            # We don't need to move the pointer
            return self.encoding[self.index + 1]

        result = -1
        # We need to move the pointer
        while True:
            # Move index by 2 (self.index is always an even number)
            self.index += 2

            if self.index >= self.m:
                # No element to exhaust
                break

            self.remaining += self.encoding[self.index]
            if self.remaining >= 0:
                # We found a value
                result = self.encoding[self.index + 1]
                break

        return result
```
