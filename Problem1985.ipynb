```python
import heapq

class StrNum:
    def __init__(self, val):
        # Note: this 'val' is string
        self.val = val


    def __lt__(self, other):
        # Primary sort
        if len(self.val) != len(other.val):
            return len(self.val) < len(other.val)

        # Secondary sort
        return self.val < other.val


class Solution:
    def kthLargestNumber(self, nums, k):
        min_heap = []

        for num in nums:
            heapq.heappush(min_heap, StrNum(num))

            if len(min_heap) > k:
                heapq.heappop(min_heap)

        return min_heap[0].val
```

```python
solution = Solution()
```

```python
nums = ["3","6","7","10"]
k = 4

print(solution.kthLargestNumber(nums, k))
```
