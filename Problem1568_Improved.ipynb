```python
class Solution:
    def minDays(self, grid):
        # Helper function: Run DFS to find articulation points and verify if it's 1 island
        def dfs(r, c, pr=-1, pc=-1):
            discovery[(r, c)] = low[(r, c)] = self.time
            self.time += 1
            children = 0

            for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
                nr, nc = r + dr, c + dc

                if 0 <= nr < rows and 0 <= nc < cols and grid[nr][nc] == 1:
                    if (nr, nc) == (pr, pc):
                        # Skip the immediate parent
                        continue

                    if (nr, nc) in discovery:
                        # Back-edge found: update low-link with discovery time of neighbour
                        low[(r, c)] = min(low[(r, c)], discovery[(nr, nc)])

                    else:
                        children += 1
                        dfs(nr, nc, r, c)

                        # Upon return, update low-link with neighbour's low-link
                        low[(r, c)] = min(low[(r, c)], low[(nr, nc)])

                        # Articulation Point Condition (Non-root)
                        if pr != -1 and low[(nr, nc)] >= discovery[(r, c)]:
                            self.has_articulation_point = True

            # Articulation Point Condition (Root of DFS tree)
            if pr == -1 and children > 1:
                self.has_articulation_point = True


        rows, cols = len(grid), len(grid[0])

        # Track discovery times and low-link values for Tarjans'
        discovery = {}
        low = {}
        self.time = 0
        self.has_articulation_point = False
        land_cells = 0

        # Conut total land cells and check starting connectivity
        # We also need a starting point for our DFS
        start_cell = None
        for r in range(rows):
            for c in range(cols):
                if grid[r][c] == 1:
                    land_cells += 1

                    if not start_cell:
                        start_cell = (r, c)

        # Base cases based on sheer land mass
        if land_cells == 0:
            return 0

        if land_cells == 1:
            return 1

        # Run the DFS from the first land cell found
        if start_cell:
            dfs(start_cell[0], start_cell[1])

        # If the DFS didn't visit all land cells, the grid was already disconnected
        if len(discovery) < land_cells:
            return 0

        # If we found an articulation point, 1 day is enoufh to split it
        if self.has_articulation_point:
            return 1

        # Otherwise, it's a strongly bound block
        return 2
```

```python
solution = Solution()
```
