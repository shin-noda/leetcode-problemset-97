```python
class Solution:
    def longestDupSubstring(self, s):
        # Idea:
        # Rabin-Karp and Binary Search

        base = 31
        mod = 10**9 + 7

        def search(length):
            # Returns the starting index of a duplicate substring of 'length', or -1
            if length == 0:
                return -1

            # Compute the hash of the very first window
            current_hash = 0
            for i in range(length):
                current_hash = (current_hash * base + ord(s[i])) % mod

            # Maps:
            #     hash -> list of starting indices
            #
            # We store all starting positions that produced the same hash.
            # If a hash collision occurs, we'll compare the actual substrings
            # before declaring a duplicate.
            seen = {current_hash: [0]}

            # This multiplier helps us quickly "delete" the leftmost character
            highest_power = pow(base, length - 1, mod)

            # Slide the window across the rest of the string
            for i in range(length, len(s)):
                # Remove the leftmost character
                first_char_val = ord(s[i - length])
                current_hash = (current_hash - first_char_val * highest_power) % mod

                # Shift left and add the rightmost character
                current_hash = (current_hash * base + ord(s[i])) % mod

                start = i - length + 1

                # Same hash found.
                if current_hash in seen:
                    # Compare against every previous substring with this hash.
                    # If one matches exactly, we found a real duplicate.
                    for prev_start in seen[current_hash]:
                        if s[prev_start:prev_start + length] == s[start:start + length]:
                            return start

                    # Hash collision. Remember this starting position too.
                    seen[current_hash].append(start)

                else:
                    # First time we've seen this hash.
                    seen[current_hash] = [start]

            return -1

        # Binary search for the maximum possible length of a duplicate substring
        left, right = 1, len(s)
        best_start = -1
        best_len = 0

        while left <= right:
            mid = (left + right) // 2
            start_idx = search(mid)

            if start_idx != -1:
                # A duplicate of length 'mid' exists. Try to find a longer one.
                best_start = start_idx
                best_len = mid
                left = mid + 1

            else:
                # Too long, no duplicaets exist. Try a shorter length
                right = mid - 1

        if best_len <= 0:
            return ""

        return s[best_start : best_start + best_len]
```

```python
solution = Solution()
```

```python
s = "banana"

print(solution.longestDupSubstring(s))
```
