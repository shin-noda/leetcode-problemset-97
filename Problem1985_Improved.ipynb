```python
import heapq

class Solution:
    def kthLargestNumber(self, nums, k):
        min_heap = []

        for num in nums:
            heapq.heappush(min_heap, (len(num), num))

            if len(min_heap) > k:
                heapq.heappop(min_heap)

        return min_heap[0][1]
```

```python
solution = Solution()
```

```python
nums = ["3","6","7","10"]
k = 4

print(solution.kthLargestNumber(nums, k))
```

```python
nums = ["2","21","12","1"]
k = 3

print(solution.kthLargestNumber(nums, k))
```
