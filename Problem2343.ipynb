```python
class Solution:
    def smallestTrimmedNumbers(self, nums, queries):
        # queries[i] = [k_i, trim_i]
        # k_i: k_i th smallest trimmed number in nums
        # trim_i: Trim each number in nums to its rightmost trim_i digits

        results = [0] * len(queries)
        trim_map = {}
        query_map = {}

        # Note: all numbers in 'nums' have the same number of digits
        min_val = 1
        max_val = len(nums[0])

        # Map the queris to 'results'
        for i, query in enumerate(queries):
            k, t = query[0], query[1]

            if t not in query_map:
                query_map[t] = {}

            if k not in query_map[t]:
                query_map[t][k] = []

            query_map[t][k].append(i)

        for i in range(min_val, max_val + 1):
            trim_map[i] = {}
            tmp = {}
            for j in range(len(nums)):
                num = int(nums[j][max_val - i:])

                if num not in tmp:
                    tmp[num] = []

                tmp[num].append(j)

            trim_map[i] = tmp

        for key in trim_map:
            if key not in query_map:
                continue

            tmp = trim_map[key]
            keys = list(tmp.keys())
            keys.sort()
            counts = 0

            for k in keys:
                bucket_size = len(tmp[k])

                for i in range(bucket_size):
                    counts += 1

                    if counts in query_map[key]:
                        for r in query_map[key][counts]:
                            results[r] = tmp[k][i]

        return results
```

```python
solution = Solution()
```

```python
nums = ["102","473","251","814"]
queries = [[1,1],[2,3],[4,2],[1,2]]

print(solution.smallestTrimmedNumbers(nums, queries))
```
