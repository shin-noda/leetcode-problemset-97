```python
import heapq

class WordFreq:
    def __init__(self, count, word):
        self.count = count
        self.word = word


    # lt -> "Less Than"
    def __lt__(self, other):
        # Min-heap behavior.
        # Primary sort: Lowest frequency comes first
        if self.count != other.count:
            return self.count < other.count

        # Secondary sort: Higher alphabetical order comes first
        return self.word > other.word


class Solution:
    def topKFrequent(self, words, k):
        # Time Complexity: O(n log(k))

        # Count frequencies manually using a standard dict
        word_map = {}
        for w in words:
            word_map[w] = word_map.get(w, 0) + 1

        # Maintain a min-heap of size k -> O(n log k)
        min_heap = []
        for word, count in word_map.items():
            heapq.heappush(min_heap, WordFreq(count, word))

            if len(min_heap) > k:
                heapq.heappop(min_heap)

        # Pop elements out and reverse the final result
        result = []
        while min_heap:
            result.append(heapq.heappop(min_heap).word)

        return result[::-1]
```

```python
solution = Solution()
```

```python
words = ["i","love","leetcode","i","love","coding"]
k = 2

print(solution.topKFrequent(words, k))
```
