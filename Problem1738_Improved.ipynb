```python
import random

class Solution:
    def kthLargestValue(self, matrix, k):
        rows, cols = len(matrix), len(matrix[0])
        n = rows * cols
        xor_arr = [0] * n

        # Populate the XOR matrix
        for r in range(rows):
            for c in range(cols):
                index = (r * cols) + c
                val = matrix[r][c]

                if c != 0:
                    val ^= xor_arr[index - 1]

                if r != 0:
                    val ^= xor_arr[index - cols]

                if c != 0 and r != 0:
                    val ^= xor_arr[index - cols - 1]

                xor_arr[index] = val

        # Iterative 3-way Quickselect
        target = k - 1
        left, right = 0, n - 1

        while left <= right:
            if left == right:
                return xor_arr[left]

            # Pick a random pivot and swapt it to the front
            pivot_idx = random.randint(left, right)
            pivot = xor_arr[pivot_idx]
            xor_arr[left], xor_arr[pivot_idx] = xor_arr[pivot_idx], xor_arr[left]

            lt = left
            gt = right
            i = left + 1

            # The Dutch National Flag 3-way scan (sorting descending)
            while i <= gt:
                if xor_arr[i] > pivot:
                    xor_arr[lt], xor_arr[i] = xor_arr[i], xor_arr[lt]
                    lt += 1
                    i += 1

                elif xor_arr[i] < pivot:
                    xor_arr[gt], xor_arr[i] = xor_arr[i], xor_arr[gt]
                    gt -= 1

                else:
                    i += 1

            # Check which bucket our target index falls into
            if lt <= target <= gt:
                # The target is inside the duplicate "Equal" bucket
                return xor_arr[target]

            elif target < lt:
                # The target is in the "Greater than" bucket on the left
                right = lt - 1

            else:
                # The target is in the "Less than" bucket on the right
                left = gt + 1

        return xor_arr[target]
```

```python
solution = Solution()
```

```python
matrix = [[5,2],[1,6]]
k = 1

print(solution.kthLargestValue(matrix, k))
```
