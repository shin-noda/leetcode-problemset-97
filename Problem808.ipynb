```python
class Solution:
    def soupServings(self, n):
        # Short circuit
        THRESHOLD = 5000
        if n > THRESHOLD:
            return 1.0

        # Scale down by 25
        # (n + 24) // 25 is a clean way to perform ceiling division in Python
        scaled_n = (n + 24) // 25

        memo = {}

        def dfs(a, b):
            # Base cases
            if a <= 0 and b <= 0:
                # Both empty
                return 0.5

            if a <= 0:
                # A empties first
                return 1.0

            if b <= 0:
                # B empties first
                return 0.0

            if (a, b) in memo:
                return memo[(a, b)]

            # Explore all 4 operations
            op1 = dfs(a - 4, b)
            op2 = dfs(a - 3, b - 1)
            op3 = dfs(a - 2, b - 2)
            op4 = dfs(a - 1, b - 3)

            memo[(a, b)] = 0.25 * (op1 + op2 + op3 + op4)

            return memo[(a, b)]

        return dfs(scaled_n, scaled_n)
```
