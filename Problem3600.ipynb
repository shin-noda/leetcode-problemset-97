```python
import heapq

class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))


    def find(self, i):
        if self.parent[i] == i:
            return i

        # Path compression
        return self.find(self.parent[i])


    def union(self, i, j):
        root_i = self.find(i)
        root_j = self.find(j)

        self.parent[root_i] = root_j


class Solution:
    def maxStability(self, n, edges, k):
        # Use Kruskal's Algorithm

        uf = UnionFind(n)
        optional_edges = []
        node_set = set()
        all_set = set()
        num_edges = 0
        min_heap = []

        for u, v, s, must in edges:
            all_set.add(u)
            all_set.add(v)
            num_edges += must

            if must == 1:
                heapq.heappush(min_heap, (s, 1))
                node_set.add(u)
                node_set.add(v)

                # Cycle detection
                if uf.find(u) == uf.find(v):
                    return -1

                # Connect 'u' and 'v'
                uf.union(u, v)

            else:
                optional_edges.append((u, v, s))

        # Check if we can form a MST
        if len(all_set) != n:
            return -1

        # Sort 'optional_edges' with the strength of the edge, decreasing order
        optional_edges.sort(key=lambda x: x[2], reverse=True)

        for u, v, s in optional_edges:
            # Cycle detection
            if uf.find(u) == uf.find(v):
                continue

            # Connect 'u' and 'v'
            uf.union(u, v)

            heapq.heappush(min_heap, (s, 0))
            num_edges += 1

        # We cannot form a spanning tree
        if num_edges != n - 1:
            return -1

        # Upgrade strength
        while k > 0 and min_heap[0][1] == 0:
            s, _ = heapq.heappop(min_heap)

            # Double the strength and marked as updated which is '1'
            heapq.heappush(min_heap, (s * 2, 1))
            k -= 1

        return min_heap[0][0]
```

```python
solution = Solution()
```

```python
n = 3
edges = [[0,1,2,1],[1,2,3,0]]
k = 1

# Expected: 2
print(solution.maxStability(n, edges, k))
```
