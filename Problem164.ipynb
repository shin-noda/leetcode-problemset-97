```python
class Solution:
    def maximumGap(self, nums):
        def radixSort(arr):
            if not arr:
                return arr

            # Helper: counting sort by digit (exp = 1, 10, 100, ...)
            def countingSortByDigit(arr, exp):
                n = len(arr)
                output = [0] * n

                # digits 0-9
                count = [0] * 10

                # Count occurrences of each digit
                for num in arr:
                    digit = (num // exp) % 10
                    count[digit] += 1

                # Convert to prefix sums
                for i in range(1, 10):
                    count[i] += count[i - 1]

                # Build output array (stable sort -> go backwards)
                for i in range(n - 1, -1, -1):
                    digit = (arr[i] // exp) % 10
                    output[count[digit] - 1] = arr[i]
                    count[digit] -= 1

                return output

            # Find max number to know number of digits
            max_val = max(arr)

            exp = 1
            while (max_val // exp) > 0:
                arr = countingSortByDigit(arr, exp)
                exp *= 10

            return arr


        # If the array contains less than two elements, return 0.
        if len(nums) < 2:
            return 0

        # Constraints: Linear time and uses linear extra space
        # The list size is bounded by <= 10^9
        # Use Radix Sort because it's O(k*N) = O(10^9 * N) = O(N)
        sorted = radixSort(nums)
        size = len(sorted)
        max_diff = 0

        for i in range(size - 1):
            max_diff = max(max_diff, sorted[i + 1] - sorted[i])

        return max_diff
```

```python
solution = Solution()
```

```python
nums = [3,6,9,1]

print(solution.maximumGap(nums))
```

```python
nums = [10]

print(solution.maximumGap(nums))
```
