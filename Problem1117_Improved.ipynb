```python
import threading

class H2O:
    def __init__(self):
        self.h_sem = threading.Semaphore(2)
        self.o_sem = threading.Semaphore(1)

        # Cyclic barrier requiring exactly 3 threads to pass together
        self.barrier = threading.Barrier(3)


    def hydrogen(self, releaseHydrogen: 'Callable[[], None]') -> None:
        self.h_sem.acquire()

        # Wait for 2 H's and 1 O to gather at the barrier
        self.barrier.wait()

        # releaseHydrogen() outputs "H". Do not change or remove this line.
        releaseHydrogen()

        # Replenish 1 slot for the next H
        # Note: we don't have to release twice because each thread replenish itself
        self.h_sem.release()



    def oxygen(self, releaseOxygen: 'Callable[[], None]') -> None:
        self.o_sem.acquire()

        # Wait for 2 H's and 1 O to gather at the barrier
        self.barrier.wait()

        # releaseOxygen() outputs "O". Do not change or remove this line.
        releaseOxygen()

        # Replenish 1 slot for the next O
        self.o_sem.release()
```
