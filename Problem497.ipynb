```python
import random

class Solution:

    def __init__(self, rects):
        # This is a rejection sampling problem.
        # Maybe not if we pick a point from valid samples.

        # We got the Wrong Answer, so it is a Rejection Sampling problem :(

        # We got the TLE for the rejection sampling solution :(((

        # Time to use the reservoir sampling solution.

        # Nope :)

        # We can do something better.
        # First, calculate the area of each rectangle.
        # e.g. r1 = 4, r2 = 18, etc.
        # Then, add them up.
        # Now, we map it like [1,4] -> r1, [5, 22] -> r2, etc.
        # Finally, let's say you get 3, it's gonna be:
        # (1, 1) <- 1, (1, 2) <- 2, (2, 1) <- 3
        # So we found the answer, (2,1)

        self.rects = rects
        self.rect_range = []
        self.min_val = 0
        self.max_val = 1

        start = 0
        end = -1
        for a, b, x, y in self.rects:
            area = (x - a + 1) * (y - b + 1)

            start = end + 1
            end = start + area - 1

            self.rect_range.append([start, end])
            self.max_val = end


    def binarySearch(self, val):
        left = 0
        right = len(self.rect_range) - 1

        while left < right:
            mid = (left + right) // 2

            if self.rect_range[mid][1] <= val:
                left = mid + 1

            else:
                right = mid - 1


        return left


    def pick(self):
        val = random.randint(self.min_val, self.max_val)
        index = self.binarySearch(val)

        if val < self.rect_range[index][0]:
            index -= 1

        elif val > self.rect_range[index][1]:
            index += 1

        rect = self.rects[index]

        # width
        cols = rect[2] - rect[0] + 1

        # Normalize the value
        val -= self.rect_range[index][0]

        x = rect[0] + (val % cols)
        y = rect[1] + (val // cols)

        return [x, y]
```
