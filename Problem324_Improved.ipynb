```python
import random

class Solution:
    def wiggleSort(self, nums):
        """
        Do not return anything, modify nums in-place instead.
        """

        # Use Quickselect + the Dutch National Flag (DNF) algorithm
        # to achieve O(2N) time complexity which is O(N)

        n = len(nums)
        if n <= 1:
            return

        # O(N) Quickselect to find median
        def quickselect(left, right, k):
            if left == right:
                return nums[left]

            # Pick a random pivot to avoid O(N^2) worst case
            pivot_idx = random.randint(left, right)
            nums[pivot_idx], nums[right] = nums[right], nums[pivot_idx]

            # Standard partition
            pivot = nums[right]
            i = left
            for j in range(left, right):
                if nums[j] < pivot:
                    nums[i], nums[j] = nums[j], nums[i]
                    i += 1

            nums[i], nums[right] = nums[right], nums[i]

            # Is this the median position?
            if i == k:
                return nums[i]

            elif i < k:
                return quickselect(i + 1, right, k)

            else:
                return quickselect(left, i - 1, k)

        # Find the actual median value
        mid_val = quickselect(0, n - 1, n // 2)

        # The Dutch National Flag with virtual indexing
        def map_idx(i):
            # We reorder the indices so that when we partition around the median,
            # the larger values naturally flow into the peak positions (odd indices),
            # and the smaller values naturally flow into the valley positions(even indices).
            return (2 * i + 1) % (n | 1)


        left = 0
        i = 0
        right = n - 1
        # With this, we reshuffle indicies like this:
        # 0 → 1
        # 1 → 3
        # 2 → 5
        # 3 → 0
        # 4 → 2
        # 5 → 4
        # So it looks like this:
        # Virtual position:   0   1   2   3   4   5
        #                     |   |   |   |   |   |
        # Real index:         1   3   5   0   2   4
        # The 'left' indices are odd and the 'right' indices are even
        while i <= right:
            if nums[map_idx(i)] > mid_val:
                # We assign odd incides to nums
                nums[map_idx(i)], nums[map_idx(left)] = nums[map_idx(left)], nums[map_idx(i)]

                # Note: we increment both 'left' and 'i' because we've already
                # classified the element that was at 'i'.
                left += 1
                i += 1

            elif nums[map_idx(i)] < mid_val:
                # We assign even indices to nums
                nums[map_idx(i)], nums[map_idx(right)] = nums[map_idx(right)], nums[map_idx(i)]

                # Note: we do NOT increment 'i' here because we don't know anything about
                # the swapped number since right is greater than 'i'.
                right -= 1

            else:
                i += 1
```

```python
solution = Solution()
```

```python
nums = [1,5,1,1,6,4]

print(solution.wiggleSort(nums))
```
