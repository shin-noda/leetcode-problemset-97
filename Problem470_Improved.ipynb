```python
class Solution:
    def rand10(self):
        """
        :rtype: int
        """

        result = -1
        while result == -1:
            # Roll twice to get a 2-digit base-7 number
            row = rand7()
            col = rand7()

            # We use number [10,40] and reject 41 to 49.
            idx = (row - 1) * 7 + col

            # If it's in our desired range [1, 40], return it.
            if idx <= 40:
                result = ((idx - 1) % 10) + 1

        return result
```
