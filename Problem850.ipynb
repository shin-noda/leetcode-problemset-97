```python
import heapq

class Solution:
    def rectangleArea(self, rectangles):
        MOD = 10**9 + 7
        total_area = 0

        # Collect and sort all unique x-coordinates
        # Extract all x-coordinates
        x_set = set()
        for r in rectangles:
            # left edge
            x_set.add(r[0])

            # right edge
            x_set.add(r[2])

        # Sort them: we need them in order so we can process the "strips" from left to right
        x_coords = sorted(x_set)

        # Sweep across the strips
        for i in range(len(x_coords) - 1):
            # (x1, x2) represents a vertical strip of the plane
            x1, x2 = x_coords[i], x_coords[i + 1]

            # Find all y-intervals that cover this vertical strip
            y_intervals = []
            for r in rectangles:
                # Check if a rectangle fully covers the entire strip horizontallylk
                if r[0] <= x1 and r[2] >= x2:
                    y_intervals.append((r[1], r[3]))

            # Calculate union of 1D intervals
            if not y_intervals:
                continue

            y_intervals.sort()

            merged_y_length = 0
            curr_y1, curr_y2 = y_intervals[0]

            for next_y1, next_y2 in y_intervals[1:]:
                # Overlap
                if next_y1 < curr_y2:
                    curr_y2 = max(curr_y2, next_y2)

                # No overlap, add previous and start new
                else:
                    merged_y_length += curr_y2 - curr_y1
                    curr_y1, curr_y2 = next_y1, next_y2

            merged_y_length += curr_y2 - curr_y1

            total_area += (x2 - x1) * merged_y_length

        return total_area % MOD
```

```python
solution = Solution()
```

```python
rectangles = [[0,0,2,2],[1,0,2,3],[1,0,3,1]]

print(solution.rectangleArea(rectangles))
```
