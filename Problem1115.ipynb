```python
import threading

class FooBar:
    def __init__(self, n):
        self.n = n

        self.foo_done = threading.Event()
        self.bar_done = threading.Event()


    def foo(self, printFoo: 'Callable[[], None]') -> None:
        for i in range(self.n):
            # printFoo() outputs "foo". Do not change or remove this line.
            printFoo()

            # Unlock the second one
            self.bar_done.set()

            # Lock this one
            self.foo_done.clear()

            # Note: Unlike .bar(), .foo() waits here
            self.foo_done.wait()


    def bar(self, printBar: 'Callable[[], None]') -> None:
        for i in range(self.n):
            self.bar_done.wait()

            # printBar() outputs "bar". Do not change or remove this line.
            printBar()

            # Unlock the first one
            self.foo_done.set()

            # Lock this one
            self.bar_done.clear()
```
