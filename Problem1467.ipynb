```python
from math import comb

class Solution:
    def getProbability(self, balls):
        total_balls = sum(balls)
        n = total_balls // 2
        k = len(balls)

        memo = {}

        def dfs(color_idx, balls_in_box1, colors_in_box1, colors_in_box2):
            # Base case: all colors processed
            if color_idx == k:
                # If Box 1 got exactly n balls and distinct color counts match,
                # this path is a 100% success
                if balls_in_box1 == n and colors_in_box1 == colors_in_box2:
                    return 1.0

                return 0.0

            # Quick Pruning: if Box 1 already has too many balls, this path is a 0% success
            if balls_in_box1 > n:
                return 0.0

            state_key = (color_idx, balls_in_box1, colors_in_box1, colors_in_box2)
            if state_key in memo:
                return memo[state_key]

            current_ball_count = balls[color_idx]

            # Dynamic tracking of remaining balls available globally and needed for Box 1
            rem_balls_total = sum(balls[color_idx:])
            rem_balls_needed_box1 = n - balls_in_box1

            total_path_probability = 0.0

            # Try every valid split of the current color's balls
            for j in range(current_ball_count + 1):
                # Ensure we don't put more balls in Box 1 than it actually needs
                if j > rem_balls_needed_box1:
                    continue

                # Ensure Box 2 isn't being forced to take more balls than are actually left
                if (current_ball_count - j) > (rem_balls_total - rem_balls_needed_box1):
                    continue

                # Calculate the exact probability of this specific split occurring
                ways_to_choose_j = comb(current_ball_count, j)
                ways_to_choose_rest = comb(rem_balls_total - current_ball_count, rem_balls_needed_box1 - j)
                total_ways_to_fill_box1 = comb(rem_balls_total, rem_balls_needed_box1)

                local_prob = (ways_to_choose_j * ways_to_choose_rest) / total_ways_to_fill_box1

                # Determine distinct color contributions
                box1_contrib = 1 if j > 0 else 0
                box2_contrib = 1 if (current_ball_count - j) > 0 else 0

                # Multiply localo probability by downstream probability
                downstream_prob = dfs(
                    color_idx + 1,
                    balls_in_box1 + j,
                    colors_in_box1 + box1_contrib,
                    colors_in_box2 + box2_contrib
                )

                total_path_probability += local_prob * downstream_prob

            memo[state_key] = total_path_probability

            return total_path_probability

        return dfs(0, 0, 0, 0)
```

```python
solution = Solution()
```
