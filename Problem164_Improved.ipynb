```python
class Solution:
    def maximumGap(self, nums):
        # If the array contains less than two elements, return 0.
        if len(nums) < 2:
            return 0

        global_min = min(nums)
        global_max = max(nums)

        # If all elements are identical, the max gap is 0
        if global_min == global_max:
            return 0

        n = len(nums)

        # Calculate bucket size and count
        # We want the bucket size to be at least 1
        bucket_size = max(1, (global_max - global_min) // (n - 1))
        bucket_count = ((global_max - global_min) // bucket_size) + 1

        # Buckets to store (min_val, max_val) for each bucket
        # Initialized to float('inf') and float('-inf')
        buckets_min = [float('inf')] * bucket_count
        buckets_max = [float('-inf')] * bucket_count

        # Map each number into a bucket
        for num in nums:
            idx = (num - global_min) // bucket_size
            buckets_min[idx] = min(buckets_min[idx], num)
            buckets_max[idx] = max(buckets_max[idx], num)

        # Scan buckts to find the maximum gap
        max_gap = 0
        previous_max = global_min

        for i in range(bucket_count):
            # Skip empty buckets
            if buckets_min[i] == float('inf'):
                continue

            # Gap is between current bucket's min and previous bucket's max
            max_gap = max(max_gap, buckets_min[i] - previous_max)
            previous_max = buckets_max[i]

        return max_gap
```

```python
solution = Solution()
```

```python
nums = [3,6,9,1]

print(solution.maximumGap(nums))
```
