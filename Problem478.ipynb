```python
import random

class Solution:

    def __init__(self, radius, x_center, y_center):
        self.radius = float(radius)
        self.r_2 = self.radius**2
        self.x_center = x_center
        self.y_center = y_center


    def randPoint(self):
        # The equasion of a circule is:
        # # (x - a)^2 + (y - b)^2 = r^2
        #
        # The range of x is:
        # (x - a)^2 <= r^2
        # -r <= x - a <= r
        # -r + a <= x <= r + a
        #
        # (x - a)^2 + (y - b)^2 <= r^2
        # (y - b)^2 <= r^2 - (x - a)^2
        # -(r^2 - (x - a)^2)^0.5 <= y - b <= (r^2 - (x - a)^2)^0.5
        # -(r^2 - (x - a)^2)^0.5 + b <= y <= (r^2 - (x - a)^2)^0.5 + b

        # You know what :O ?
        # Just do random and reject illigal points.

        x = 0.0
        y = 0.0

        is_calculating = True
        while is_calculating:
            x = random.uniform(-self.radius, self.radius) + self.x_center
            y = random.uniform(-self.radius, self.radius) + self.y_center

            # Make sure that it doesn't go beyond the radius
            if (x - self.x_center)**2 + (y - self.y_center)**2 <= self.r_2:
                is_calculating = False

        return [x, y]
```
