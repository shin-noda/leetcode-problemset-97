```python
import threading

class H2O:
    count = 0

    def __init__(self):
        self.h_sem = threading.Semaphore(2)
        self.o_sem = threading.Semaphore(1)


    def hydrogen(self, releaseHydrogen: 'Callable[[], None]') -> None:
        self.h_sem.acquire()
        H2O.count += 1

        # releaseHydrogen() outputs "H". Do not change or remove this line.
        releaseHydrogen()

        self.releaseHydrogenAndOxygen()


    def oxygen(self, releaseOxygen: 'Callable[[], None]') -> None:
        self.o_sem.acquire()
        H2O.count += 1

        # releaseOxygen() outputs "O". Do not change or remove this line.
        releaseOxygen()

        self.releaseHydrogenAndOxygen()


    def releaseHydrogenAndOxygen(self) -> None:
        if H2O.count >= 3:
            # Release both hydrogen and oxygen threads
            # Note: we need to release 'h_sem' twice to restore the count of 2.
            self.h_sem.release()
            self.h_sem.release()
            self.o_sem.release()

            # Reset the counter
            H2O.count = 0
```
