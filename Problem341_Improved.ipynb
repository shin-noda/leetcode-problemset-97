```python
class NestedIterator:
    def __init__(self, nestedList: [NestedInteger]):
        self.gen = self._generator(nestedList)
        self.peek = None


    def _generator(self, current_list):
        for item in current_list:
            if item.isInteger():
                yield item.getInteger()

            else:
                yield from self._generator(item.getList())


    def next(self) -> int:
        result = self.peek
        self.peek = None
        return result


    def hasNext(self) -> bool:
        # Note: this if block exists so that it makes hasNext() idempotent.
        # e.g. hasNext() will be called multiple times.
        # It's not required for this problem because it's guaranteed that
        # next() is guaranteed to be called after hasNext() but it's
        # a good safe-guard to have to make the program robust.
        if self.peek is not None:
            return True

        try:
            self.peek = next(self.gen)
            return True

        except StopIteration:
            return False
```
