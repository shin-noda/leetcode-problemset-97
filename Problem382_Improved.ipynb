```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
import random

class Solution:

    def __init__(self, head):
        self.n = 1
        self.head = head

        # Use the Tortoise and the Hare algorithm
        slow = head
        fast = head

        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            self.n += 1

        if not fast:
            self.n = (self.n - 1) * 2

        else:
            self.n = (self.n * 2) - 1


    def getRandom(self):
        # Generate a random value
        rand_val = random.randint(0, self.n - 1)

        head = self.head
        for _ in range(rand_val):
            head = head.next

        return head.val
```
