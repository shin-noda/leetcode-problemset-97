```python
class Solution:
    def countElements(self, nums, k):
        # Quick check
        if k == 0:
            return len(nums)

        # Sort 'nums'. O(NlogN)
        nums.sort(reverse=True)

        count = 0
        prev = nums[0]

        for num in nums:
            if num != prev:
                prev = num

                if count >= k:
                    break

            count += 1

        return len(nums) - count
```

```python
solution = Solution()
```

```python
nums = [3,1,2]
k = 1

# Expected: 2
print(solution.countElements(nums, k))
```
