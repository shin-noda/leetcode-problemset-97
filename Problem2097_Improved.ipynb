```python
class Solution:
    def validArrangement(self, pairs):
        # Build adjacency list and track degree balances
        # node_map stores: out_degree - in_degree
        adj = {}
        node_map = {}

        for src, dst in pairs:
            if src not in adj:
                adj[src] = []

            adj[src].append(dst)

            node_map[src] = node_map.get(src, 0) + 1
            node_map[dst] = node_map.get(dst, 0) - 1

        # Find the starting node
        start = pairs[0][0]
        for node, balance in node_map.items():
            if balance == 1:
                start = node
                break

        # Iterative Hierholzer's Algorithm (No recursion limit issues)
        r_path = []
        curr_stack = [start]

        while curr_stack:
            curr_node = curr_stack[-1]

            # If the current node still has outgoing edges, keep traversing
            if curr_node in adj and adj[curr_node]:
                next_node = adj[curr_node].pop()
                curr_stack.append(next_node)

            else:
                # Dead end reached: push to the reversed path and backtrack
                r_path.append(curr_stack.pop())

        # Recounstruct the pairs from the reversed path
        result = []
        path = r_path[::-1]

        for i in range(len(path) - 1):
            result.append([path[i], path[i + 1]])

        return result
```

```python
solution = Solution()
```

```python
pairs = [[5,1],[4,5],[11,9],[9,4]]

print(solution.validArrangement(pairs))
```
