```python
class PeekingIterator:
    def __init__(self, iterator):
        """
        Initialize your data structure here.
        :type iterator: Iterator
        """

        self.iterator = iterator

        self.prev_pointer = 0
        self.curr_pointer = 0
        self.prev_value = None
        self.curr_value = None

    def peek(self):
        """
        Returns the next element in the iteration without advancing the iterator.
        :rtype: int
        """

        # When 'prev' and 'curr' pointers are the same, we only change the value of 'curr'
        # so that .hasNext() of PeekingIterator returns the correct value.
        if self.prev_pointer == self.curr_pointer:
            self.curr_pointer += 1
            self.curr_value = self.iterator.next()

        return self.curr_value


    def next(self):
        """
        :rtype: int
        """

        # We only move the pointer when 'prev' and 'curr' are the same.
        if self.prev_pointer == self.curr_pointer:
            self.curr_value = self.iterator.next()
            self.curr_pointer += 1

        self.prev_value = self.curr_value
        self.prev_pointer += 1

        return self.curr_value


    def hasNext(self):
        """
        :rtype: bool
        """

        if self.prev_pointer == self.curr_pointer:
            return self.iterator.hasNext()

        # If 'prev' and 'curr' have different values, it means that 'iterator's pointer
        # is one step ahead due to .peek(), so it's guaranteed to have the next element.
        return True
```
