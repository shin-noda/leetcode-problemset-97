```python
import random

class Solution:

    def __init__(self, nums):
        # We're gonna sort the list, so we want to preserve
        # the original indices of 'nums
        paired_nums = []
        for i in range(len(nums)):
            paired_nums.append((nums[i], i))

        # O(N log N) where N = len(nums)
        paired_nums.sort()

        self.paired_nums = paired_nums

        self.index_map = {}
        for i in range(len(self.paired_nums)):
            num = self.paired_nums[i][0]

            if num not in self.index_map:
                self.index_map[num] = []

            self.index_map[num].append(i)


    def pick(self, target):
        # Use Reservoir Sampling
        # Don't use Reservoir Sampling because the TLE.

        n = len(self.index_map[target])
        rand_val = random.randint(0, n - 1)

        return self.paired_nums[self.index_map[target][rand_val]][1]
```
