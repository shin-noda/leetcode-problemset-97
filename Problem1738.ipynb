```python
import heapq

class Solution:
    def kthLargestValue(self, matrix, k):
        min_heap = []
        rows, cols = len(matrix), len(matrix[0])

        # This is a 1D matrix.
        xor_arr = [0] * (cols * rows)

        for r in range(rows):
            for c in range(cols):
                index = (r * cols) + c
                xor_arr[index] = matrix[r][c]

                if c != 0:
                    xor_arr[index] ^= xor_arr[index - 1]

                if r != 0:
                    xor_arr[index] ^= xor_arr[index - cols]

                # Avoid the double xoring
                if c != 0 and r != 0:
                    # XOR again the top-left diagonal cell to cancel double counting
                    xor_arr[index] ^= xor_arr[index - cols - 1]

                heapq.heappush(min_heap, xor_arr[index])

                if len(min_heap) > k:
                    popped = heapq.heappop(min_heap)

        return min_heap[0]
```

```python
solution = Solution()
```

```python
matrix = [[5,2],[1,6]]
k = 1

print(solution.kthLargestValue(matrix, k))
```
