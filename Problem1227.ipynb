```python
class Solution:
    def nthPersonGetsNthSeat(self, n):
        # N = 1. It's guaranteed.

        # N = 2. It's 50/50.

        # N = 3.
        # Case 1: the first passenger sits at the 3rd seat -> Failure
        # Case 2: the first passenger sits at the 1st seat -> Success.
        # Case 3: the first passenger sits at the 2nd seat:
        # -> 2nd passenger sits at the 1st seat -> Success
        # -> 2nd passenger sits at the 3rd seat -> Failure
        # 0 + 1/3 + 1/3 * (1/2 + 0) = 1/3 + 1/6 = 1/2

        # N = 4.
        # Case 1: the first passenger sits at the 4th seat -> Failure
        # Case 2: the first passenger sits at the 1st seat -> Success
        # Case 3: the first passenger sits at the 3rd seat:
        # -> 2nd passenger sits at the 2nd seat
        # ---> 3rd passenger sits at the 1st seat -> Success
        # ---> 3rd passenger sits at the 4th seat -> Failure
        #
        # Case 4: the first passenger sits at the 2nd seat:
        # -> 2nd passenger sits at the 4th seat -> Failure
        # -> 2nd passenger sits at the 1st seat -> Success
        # -> 2nd passenger sits at the 3rd seat:
        # ---> 3rd passenger sits at the 1st seat -> Success
        # ---> 3rd passenger sits at the 4th seat -> Failure
        #
        # Probabilities:
        # Case 1: 1/4 -> Failure
        # Case 2: 1/4 -> Success
        # Case 3:
        # 3-2-1-4: 1/4 * 1 * 1/2 = 1/8 -> Success
        # 3-2-4-1: 1/4 * 1 * 1/2 = 1/8 -> Failure
        #
        # Case 4:
        # 2-4: 1/4 * 1/3 = 1/12 -> Failure
        # 2-1: 1/4 * 1/3 = 1/12 -> Success
        # 2-3-1-4: 1/4 * 1/3 * 1/2 = 1/24 -> Success
        # 2-3-4-1: 1/4 * 1/3 * 1/2 = 1/24 -> Failure
        #
        # (1/4) * 0 +
        # (1/4) +
        # (1/8 + 1/8 * 0) +
        #(1/12 * 0 + 1/12 + 1/24 + 1/24 * 0)
        # = 0.5

        # Is this always 50% except n = 1 :O ???
        # Probablly :)))

        # Let's test it with n = 5.
        # N = 5.
        # Case 1: the first person picks the first seat. 1/5 -> success
        # Case 2: the first person picks the fifth seat. 1/5 * 0 -> failure
        # Case 3: the first person picks the fourth seat.
        # -> 2nd and 3rd person picks the 2nd and the 3rd seats respectively.
        # ---> 4th person picks the 1st seat -> 4-2-3-1-5: 1/5 * 1/2 -> success
        # ---> 4th person picks the 5th seat -> 4-2-3-5-1: 1/5 * 1/2 * 0 -> failure
        #
        # Case 4: the first person picks the 3rd seat.
        # -> 2nd person picks the 2nd seat.
        # ---> 3rd person picks the 1st seat: 3-2-1-4-5: 1/5 * 1/3 -> success
        # ---> 3rd person picks the 5th seat: 3-2-5: 1/5 * 1/3 * 0  -> failure
        # ---> 3rd person picks the 4th seat:
        # -----> 4th person picks the 1st seat: 3-2-4-1-5: 1/5 * 1/3 * 1/2 -> Success
        # -----> 4th person picks the 5th seat: 3-2-4-5-1: 1/5 * 1/3 * 1/2 * 0 -> Failure
        #
        # Case 5: the first person picks the 2nd seat.
        # -> 2nd person picks the 1st seat: 2-1: 1/5 * 1/4 -> success
        # -> 2nd person picks the 5th seat: 2-1: 1/5 * 1/4 * 0 -> failure
        # -> 2nd person picks the 4th seat:
        # ---> 3rd person picks the 3rd seat.
        # -----> 4th person picks the 1st seat: 2-4-3-1-5: 1/5*1/4*1/2 -> success
        # -----> 4th person picks the 5th seat: 2-4-3-5-1: 1/5*1/4*1/2 * 0 -> failure
        # -> 2nd person picks the 3rd seat:
        # ---> 3rd person picks the 1st seat: 2-3-1: 1/5*1/4*1/3 -> success
        # ---> 3rd person picks the 5th seat: 2-3-5: 1/5*1/4*1/3 * 0 -> failure
        # ---> 3rd person picks the 4th seat: 2-3-4:
        # -----> 4th person picks the 1st seat: 2-3-4-1-5: 1/5*1/4*1/3*1/2 success
        # -----> 4th person picks the 5th seat: 2-3-4-5-1: 1/5*1/4*1/3*1/2 * 0 failure
        #
        # The probability is:
        # (
        #     (1/5)+
        #     (1/5) * 0+
        #     (1/5*1/2 + 1/5*1/2 * 0)+
        #     (1/5*1/3 + 1/5*1/3 * 0 + 1/5*1/3*1/2 + 1/5*1/3*1/2 * 0) +
        #     (
        #         1/5*1/4 +
        #         1/5*1/4 * 0 +
        #         1/5*1/4*1/2 +
        #         1/5*1/4*1/2 * 0 +
        #         1/5*1/4*1/3 +
        #         1/5*1/4*1/3 * 0 +
        #         1/5*1/4*1/3*1/2 +
        #         1/5*1/4*1/3*1/2 * 0
        #     )
        # ) = 0.5

        # Yuuuuuuuup. It's probabily 0.5 :)))

        if n == 1:
            return 1.0

        return 0.5
```

```python
# Case 1: the first person picks the first seat. 1/5 -> success
# Case 2: the first person picks the fifth seat. 1/5 * 0 -> failure
# Case 3: the first person picks the fourth seat.
# -> 2nd and 3rd person picks the 2nd and the 3rd seats respectively.
# ---> 4th person picks the 1st seat -> 4-2-3-1-5: 1/5 * 1/2 -> success
# ---> 4th person picks the 5th seat -> 4-2-3-5-1: 1/5 * 1/2 * 0 -> failure
#
# Case 4: the first person picks the 3rd seat.
# -> 2nd person picks the 2nd seat.
# ---> 3rd person picks the 1st seat: 3-2-1-4-5: 1/5 * 1/3 -> success
# ---> 3rd person picks the 5th seat: 3-2-5: 1/5 * 1/3 * 0  -> failure
# ---> 3rd person picks the 4th seat:
# -----> 4th person picks the 1st seat: 3-2-4-1-5: 1/5 * 1/3 * 1/2 -> Success
# -----> 4th person picks the 5th seat: 3-2-4-5-1: 1/5 * 1/3 * 1/2 * 0 -> Failure
#
# Case 5: the first person picks the 2nd seat.
# -> 2nd person picks the 1st seat: 2-1: 1/5 * 1/4 -> success
# -> 2nd person picks the 5th seat: 2-1: 1/5 * 1/4 * 0 -> failure
# -> 2nd person picks the 4th seat:
# ---> 3rd person picks the 3rd seat.
# -----> 4th person picks the 1st seat: 2-4-3-1-5: 1/5*1/4*1/2 -> success
# -----> 4th person picks the 5th seat: 2-4-3-5-1: 1/5*1/4*1/2 * 0 -> failure
# -> 2nd person picks the 3rd seat:
# ---> 3rd person picks the 1st seat: 2-3-1: 1/5*1/4*1/3 -> success
# ---> 3rd person picks the 5th seat: 2-3-5: 1/5*1/4*1/3 * 0 -> failure
# ---> 3rd person picks the 4th seat: 2-3-4:
# -----> 4th person picks the 1st seat: 2-3-4-1-5: 1/5*1/4*1/3*1/2 success
# -----> 4th person picks the 5th seat: 2-3-4-5-1: 1/5*1/4*1/3*1/2 * 0 failure

(
    (1/5)+
    (1/5) * 0+
    (1/5*1/2 + 1/5*1/2 * 0)+
    (1/5*1/3 + 1/5*1/3 * 0 + 1/5*1/3*1/2 + 1/5*1/3*1/2 * 0) +
    (
        1/5*1/4 +
        1/5*1/4 * 0 +
        1/5*1/4*1/2 +
        1/5*1/4*1/2 * 0 +
        1/5*1/4*1/3 +
        1/5*1/4*1/3 * 0 +
        1/5*1/4*1/3*1/2 +
        1/5*1/4*1/3*1/2 * 0
    )
)
```
