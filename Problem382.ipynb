```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
import random

class Solution:

    def __init__(self, head):
        self.node_vals = []
        self.n = 0

        while head:
            self.node_vals.append(head.val)
            self.n += 1

            # Proceed to the next node
            head = head.next


    def getRandom(self):
        return self.node_vals[random.randint(0, self.n - 1)]
```
