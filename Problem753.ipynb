```python
class Solution:
    def crackSafe(self, n, k):
        # This is a de Bruijin sequestion.
        # Use Hierholzer's Algorithm.

        # If n = 1, we just need to output all digits from 0 to k - 1
        if n == 1:
            return "".join(str(i) for i in range(k))

        visited = set()
        path = []

        def dfs(node):
            for x in range(k):
                # Form the full n-digit password candidate
                neighbour = node + str(x)

                if neighbour not in visited:
                    visited.add(neighbour)

                    # Move to the next node (the last n - 1 digits of our candidate)
                    dfs(neighbour[1:])

                    # Hierholzer's step: add the edge digit to our path post-order
                    path.append(str(x))

        # Start node is (n - 1) zeros
        start_node = "0" * (n - 1)
        dfs(start_node)

        # The path gives us the sequence in reverse.
        # We append the start_node at the end because Hierholzer reconstructs backwards
        # then re reverse it.
        return start_node +  "".join(path[::-1])
```

```python
solution = Solution()
```

```python
n = 1
k = 1

print(solution.crackSafe(n, k))
```
