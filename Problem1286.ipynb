```python
class CombinationIterator:

    def __init__(self, characters: str, combinationLength: int):
        # Idea: Backtracking + Generator
        self.characters = characters
        self.combinationLength = combinationLength

        self.tmp = []
        self.gen = self._generator(0, combinationLength)

        # Generate the first item
        self.peek = next(self.gen)



    def _generator(self, index, remaining):
        # Base Case
        if remaining <= 0:
            yield "".join(self.tmp)
            return

        for i in range(index, len(self.characters)):
            # Choose
            self.tmp.append(self.characters[i])

            # Explore
            yield from self._generator(i + 1, remaining - 1)

            # Unchoose
            self.tmp.pop()


    def next(self) -> str:
        try:
            result = self.peek
            self.peek = next(self.gen)

        except StopIteration:
            self.peek = None

        return result


    def hasNext(self) -> bool:
        if self.peek is not None:
            return True

        return False
```
