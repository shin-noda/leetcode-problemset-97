```python
class Solution:
    def frequencySort(self, s):
        freq_map = {}

        for ch in s:
            freq_map[ch] = freq_map.get(ch, 0) + 1

        items = list(freq_map.items())

        # O(N log N)
        items.sort(key=lambda x: x[1], reverse=True)

        result = []
        for item in items:
            result.extend(item[0] * item[1])

        return "".join(result)
```

```python
solution = Solution()
```

```python
s = "tree"

print(solution.frequencySort(s))
```

```python
s = "cccaaa"

print(solution.frequencySort(s))
```
