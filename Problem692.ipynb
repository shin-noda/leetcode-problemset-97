```python
import heapq

class Solution:
    def topKFrequent(self, words, k):
        # Time Complexity: O(n + klog(m))

        max_heap = []
        word_map = {}
        for w in words:
            word_map[w] = word_map.get(w, 0) + 1

        for key in word_map:
            count = word_map[key]
            max_heap.append((-count, key))

        # O(m) where m is the number of elements currently in 'max_heap'.
        heapq.heapify(max_heap)

        result = []

        # O(klog(m))
        for _ in range(k):
            _, word = heapq.heappop(max_heap)
            result.append(word)

        return result
```

```python
solution = Solution()
```

```python
words = ["i","love","leetcode","i","love","coding"]
k = 2

print(solution.topKFrequent(words, k))
```
