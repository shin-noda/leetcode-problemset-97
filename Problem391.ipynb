```python
import heapq

class Solution:
    def isRectangleCover(self, rectangles):
        area = 0
        x1, y1 = float('inf'), float('inf')
        x2, y2 = float('-inf'), float('-inf')
        corners = set()

        for x, y, a, b in rectangles:
            # Accumulate total area
            area += (a - x) * (b - y)

            # Track bounding box
            x1, y1 = min(x1, x), min(y1, y)
            x2, y2 = max(x2, a), max(y2, b)

            # Toggle corners in the set
            for corner in [(x, y), (x, b), (a, y), (a, b)]:
                if corner in corners:
                    corners.remove(corner)

                else:
                    corners.add(corner)

        # Check results
        # Bounding box must have eaxtly 4 corners
        if (
            len(corners) != 4 or
            (x1, y1) not in corners or (x1, y2) not in corners or
            (x2, y1) not in corners or (x2, y2) not in corners
        ):
            return False

        # Area must match exactly
        return area == (x2 - x1) * (y2 - y1)
```

```python
solution = Solution()
```

```python
rectangles = [[1,1,3,3],[3,1,4,2],[3,2,4,4],[1,3,2,4],[2,3,3,4]]

print(solution.isRectangleCover(rectangles))
```
