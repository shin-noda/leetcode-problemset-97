```python
import random

class Solution:

    def __init__(self, m, n):
        # Idea:
        # I got Memory Limit Exceeded.
        # Time to not to use a matrix.

        self.m = m
        self.n = n

        self.visited = set()

        self.min_val = 0
        self.max_val = m * n - 1


    def convertNumToRC(self, num):
        col = num % self.n
        row = num // self.n

        return (row, col)


    def flip(self):
        while True:
            rand_val = random.randint(self.min_val, self.max_val)
            r, c = self.convertNumToRC(rand_val)

            if (r, c) not in self.visited:
                self.visited.add((r, c))
                break

        return [r, c]


    def reset(self):
        self.visited = set()
```
