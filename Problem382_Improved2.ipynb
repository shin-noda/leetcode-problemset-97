```python
# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
```

```python
import random

class Solution:

    def __init__(self, head):
        # We only store the head, so O(1) space.
        self.head = head


    def getRandom(self):
        result = self.head.val
        curr = self.head
        i = 1

        while curr:
            # Generate a random integer between 1 and i (inclusive)
            # There is a exactly a 1 / i chance that random.randint(1, i) == 1
            if random.randint(1, i) == 1:
                result = curr.val

            curr = curr.next
            i += 1

        return result
```
