```python
class Solution:
    def intervalIntersection(self, firstList, secondList):
        result = []
        n1 = len(firstList)
        n2 = len(secondList)

        i1 = i2 = 0
        while i1 < n1 and i2 < n2:
            s1, e1 = firstList[i1]
            s2, e2 = secondList[i2]

            # Check overlaps
            if s1 <= s2 <= e1 or s1 <= e2 <= e1 or s2 <= s1 <= e2 or s2 <= e1 <= e2:
                # [left_end, right_end]
                result.append([max(s1, s2), min(e1, e2)])

            if e1 < e2:
                # Proceed 'firstList'
                i1 += 1

            elif e2 < e1:
                # Proceed 'secondList'
                i2 += 1

            else:
                # Proceed both
                i1 += 1
                i2 += 1

        return result
```

```python
solution = Solution()
```

```python
firstList = [[0,2],[5,10],[13,23],[24,25]]
secondList = [[1,5],[8,12],[15,24],[25,26]]

print(solution.intervalIntersection(firstList, secondList))
```
