```python
import threading

class DiningPhilosophers:
    def __init__(self):
        # Only one philosopher is allowed to pick one at a time to avoid a deadlock.
        self.picking = threading.Semaphore(1)

        self.dining = threading.Semaphore(5)
        self.barrier = threading.Barrier(5)


    # call the functions directly to execute, for example, eat()
    def wantsToEat(self,
                   philosopher: int,
                   pickLeftFork: 'Callable[[], None]',
                   pickRightFork: 'Callable[[], None]',
                   eat: 'Callable[[], None]',
                   putLeftFork: 'Callable[[], None]',
                   putRightFork: 'Callable[[], None]') -> None:

        # Only one philosopher can pick forks at the moment so that avoid a deadlock.
        # When each philosopher takes a left or right fork to them, it'll cause a deadlock.
        self.picking.acquire()

        # After successfully acuiring the picking semaphore, the philosopher starts dining
        self.dining.acquire()

        # Eating sequence
        pickLeftFork()
        pickRightFork()
        eat()
        putLeftFork()
        putRightFork()

        # After eating, the philosopher release the picking semaphore.
        self.picking.release()

        # Wait until other four philosophers finish eating
        self.barrier.wait()

        # Replenish 1 slot for dining for this philosopher
        self.dining.release()
```
