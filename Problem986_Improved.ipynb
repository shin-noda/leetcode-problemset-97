```python
class Solution:
    def intervalIntersection(self, firstList, secondList):
        result = []
        n1 = len(firstList)
        n2 = len(secondList)

        i1 = i2 = 0
        while i1 < n1 and i2 < n2:
            s1, e1 = firstList[i1]
            s2, e2 = secondList[i2]

            # Find the boundaries of the potential intersection
            start = max(s1, s2)
            end = min(e1, e2)

            # If it's a valid interval, add it to the results
            if start <= end:
                result.append([start, end])

            # Move the pointer that finishes earlier
            if e1 < e2:
                i1 += 1

            else:
                # This safely handles both e2 < e1 and e1 == e2
                i2 += 1

        return result
```

```python
solution = Solution()
```

```python
firstList = [[0,2],[5,10],[13,23],[24,25]]
secondList = [[1,5],[8,12],[15,24],[25,26]]

print(solution.intervalIntersection(firstList, secondList))
```
