```python
class Solution:
    def sampleStats(self, count):
        # The actual challenge for this problem is how to calculate
        # median efficiently. That's the entire problem.

        # Not sure if I can do it with one-pass.
        # If it's not one pass, it's not hard.

        min_val = float('inf')
        max_val = float('-inf')
        meanc = 0.0
        median = 0.0
        mode = 0.0

        sum_val = 0
        total_count = 0
        max_freq = 0
        indices = []

        for num, count in enumerate(count):
            # If count is 0, it simply adds 0.
            sum_val += num * count
            total_count += count

            if count != 0:
                # Always update the 'max_val' if count != 0
                max_val = num

                # Only update once in the entire loop
                if min_val == float('inf'):
                    min_val = num

                if count > max_freq:
                    mode = num
                    max_freq = count

                indices.append((total_count, num))

        # Calculate the mean
        mean = sum_val / float(total_count)

        #Calculate the median
        median_indices = [total_count // 2 + 1]
        if total_count % 2 ==0:
            median_indices.insert(0, total_count // 2)

        # Binary search
        left = 0
        right = len(indices) - 1
        while left <= right:
            mid = (left + right) // 2

            if indices[mid][0] >= median_indices[0]:
                right = mid - 1

            else:
                left = mid + 1

        # There is one median index
        if len(median_indices) == 1:
            median = indices[left][1]

        # There are two median indices
        else:
            med0 = indices[left][1]

            if (left == len(indices) - 1) or (median_indices[1] <= indices[left][0]):
                med1 = indices[left][1]

            else:
                med1 = indices[left + 1][1]

            median = (med0 + med1) / 2.0

        return [min_val, max_val, mean, median, mode]
```
