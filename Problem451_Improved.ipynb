```python
class Solution:
    def frequencySort(self, s):
        # Count frequencies
        freq_map = {}
        for ch in s:
            freq_map[ch] = freq_map.get(ch, 0) + 1

        # Create buckets where index = frequency
        # Max possible frequency is len(s)
        buckets = [[] for _ in range(len(s) + 1)]
        for ch, count in freq_map.items():
            buckets[count].append(ch)

        # Build the result from highest frequency to lowest
        result = []
        for freq in range(len(buckets) - 1, 0, -1):
            for ch in buckets[freq]:
                result.append(ch * freq)

        return "".join(result)
```

```python
solution = Solution()
```

```python
s = "tree"

print(solution.frequencySort(s))
```
