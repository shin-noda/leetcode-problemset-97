```python
class Solution:
    def sampleStats(self, count):
        # Get total count and basic statistics
        # Since len(count) <= 256, sum(count) is O(1), not O(N)
        total_count = sum(count)

        # Find min, max, mode, and total sum
        min_val = next(i for i, c in enumerate(count) if c > 0)
        max_val = next(i for i in range(255, -1, -1) if count[i] > 0)
        mode = max(range(256), key=lambda x: count[x])
        mean = sum(num * c for num, c in enumerate(count)) / total_count

        # Find the median elements using a running count
        # We need the elmentsat these cumulative counts (1-indexed)
        mid1_idx = (total_count + 1) // 2
        mid2_idx = (total_count // 2) + 1

        med1 = med2 = None
        current_cumulative = 0

        for num, c in enumerate(count):
            current_cumulative += c

            if med1 is None and current_cumulative >= mid1_idx:
                med1 = num

            if med2 is None and current_cumulative >= mid2_idx:
                med2 = num
                break

        median = (med1 + med2) / 2.0

        return [float(min_val), float(max_val), mean, median, float(mode)]
```
