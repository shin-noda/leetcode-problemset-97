```python
class NestedIterator:
    def __init__(self, nestedList: [NestedInteger]):
        # Idea: pre-process all the list.
        self.nestedList = nestedList
        self.iterator = []
        self.pointer = 0

        # Flatten NestedInteger
        self.preprocessNestedList(self.nestedList)

        # Assign the length of 'iterator'
        self.n = len(self.iterator)


    def next(self) -> int:
        result = self.iterator[self.pointer]
        self.pointer += 1
        return result


    def hasNext(self) -> bool:
        return self.pointer <= self.n - 1


    def preprocessNestedList(self, nested):
        m = len(nested)

        for i in range(m):
            # Check if it's a list or not
            if nested[i].isInteger():
                self.iterator.append(nested[i].getInteger())

            else:
                # Recursively call preprocessNestedList()
                self.preprocessNestedList(nested[i].getList())
```
