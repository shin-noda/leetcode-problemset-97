```python
class Solution:
    def wiggleSort(self, nums):
        """
        Do not return anything, modify nums in-place instead.
        """

        # Sort the array
        nums.sort()

        # Quick check
        if len(nums) <= 2:
            return

        # Split sorted 'nums' into first half and the second half.
        # If len(nums) is odd, the first half has one more element
        small_arr = []
        big_arr =[]

        half = len(nums) // 2
        if len(nums) % 2 == 1:
            half += 1

        small_arr = nums[:half]
        big_arr = nums[half:]

        # Sort small_arr and big_arr in reverse order
        small_arr.reverse()
        big_arr.reverse()

        for i in range(len(small_arr)):
            nums[2 * i] = small_arr[i]

        for i in range(len(big_arr)):
            nums[2 * i + 1] = big_arr[i]
```

```python
solution = Solution()
```

```python
nums = [1,5,1,1,6,4]

print(solution.wiggleSort(nums))
```
