```python
class Solution:
    def containsNearbyAlmostDuplicate(self, nums, indexDiff, valueDiff):
        # Key Insight
        # We only store at most one value per bucket because:
        # - If a bucket is empty -> we store it.
        # - If a bucket has already has an element -> we return True
        # so when we check neighbours like:
        # - abs(num - buckets[bucket_id - 1]) <= valueDiff or
        # - abs(num - buckets[bucket_id + 1]) <= valueDiff
        # we are only checking one value each because
        # that's all we nned to check.
        #
        # If two numbers differ by <= 'valueDiff', they must be either
        # - in the same bucket, or
        # - in adjacent buckets (bucket_id - 1 or bucket_id + 1)

        buckets = {}

        # The width of our bucket
        # Note: we make buckts slightly larger than the allowed difference so
        # that same-bucket guarantees validity.
        w = valueDiff + 1

        for i, num in enumerate(nums):
            # Identify which bucket 'num' belongs to
            # e.g., if w = 4, then 0, 1, 2, 3 -> bucket 0.
            # 4, 5, 6, 7 -> bucket 1.
            # In Python, -1 // 4 is -1.
            # In general,
            # bucket -2: [-2w, -w - 1]
            # bucket -1: [-w, -1]
            # bucket 0: [0, w - 1]
            # bucket 1: [w, 2w - 1]
            # bucket 2: [2w, 3w - 1]
            bucket_id = num // w

            # Condition 1: Someone is already in this bucket
            if bucket_id in buckets:
                return True

            # Condition 2: Check the left neighbouring bucket
            if (
                bucket_id - 1 in buckets and
                abs(num - buckets[bucket_id - 1]) <= valueDiff
            ):
                return True

            # Condition 3: Check the right neighbouring bucket
            if (
                bucket_id + 1 in buckets and
                abs(num - buckets[bucket_id + 1]) <= valueDiff
            ):
                return True

            # Add current number to its bucket
            buckets[bucket_id] = num

            # Sliding Window: Remove the element that falls out of 'indexDiff' boundary
            if i >= indexDiff:
                # Calculate the bucket ID of the element leaving our window
                old_bucket_id = nums[i - indexDiff] // w
                buckets.pop(old_bucket_id)

        return False
```

```python
solution = Solution()
```

```python
nums = [1,2,3,1]
indexDiff = 3
valueDiff = 0

# Expected: True
print(solution.containsNearbyAlmostDuplicate(nums, indexDiff, valueDiff))
```

```python
nums = [1,5,9,1,5,9]
indexDiff = 2
valueDiff = 3

# Expected: False
print(solution.containsNearbyAlmostDuplicate(nums, indexDiff, valueDiff))
```
